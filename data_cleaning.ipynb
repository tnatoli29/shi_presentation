```python
#Import your dependences 
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
```

```python
#Select data from CSV file and convert them into Panda DataFrames
july_2016 = "Resources/July2016_citibike_JC.csv"
july_2016_df = pd.read_csv(july_2016)

july_2019 = "Resources/July2019_citibike_JC.csv"
july_2019_df = pd.read_csv(july_2019)

#Convert all column names to lower case to match 2019 columns
july_2016_df.columns= july_2016_df.columns.str.lower()

#Show the data frame, using the head function to limit the row to 10. 
july_2016_df.head(10)
```

```python
#Changing gender column to male, female, or unknown due to the records having it as a numerical categories of 0, 1, 2 
#First use the map function to convert the numerical column into a string
july_2016_df["gender"] = july_2016_df["gender"].map(str)

#Now convert the values of 0, 1, 2 into their gender values 
july_2016_df["gender"] = july_2016_df["gender"].replace(['0','1','2'],['Unknown','Male','Female'])

#Show the DF to make sure the changes took place
july_2016_df.head(10)
```

```python
#Make the following changes above to the 2019 DF 
july_2019_df["gender"] = july_2019_df["gender"].map(str)
july_2019_df["gender"] = july_2019_df["gender"].replace(['0','1','2'],['Unknown','Male','Female'])
july_2019_df.head(10)
```

```python
#Next we need to split starttime/end time columns to make them their own columns 
#This first step will make sure all the values in the Start Time column are in the correct format 
july_2016_df['start time'] = pd.to_datetime(july_2016_df['start time'])

#Next we will take the date portion of start time and create a new column with just the date 
july_2016_df['start_date'] = july_2016_df['start time'].dt.date

#Now take the time of create its own column 
july_2016_df['start_time'] = july_2016_df['start time'].dt.time

#Finally remove the old column since it is no longer needed
del july_2016_df['start time']

july_2016_df.head(10)
```

```python
#We will do the do the same steps as before to the Endtime column in 2016 and the start and stop times of the 2019 DF
#2016 Stoptimes
july_2016_df['stop time'] = pd.to_datetime(july_2016_df['stop time'])
july_2016_df['end_date'] = july_2016_df['stop time'].dt.date
july_2016_df['end_time'] = july_2016_df['stop time'].dt.time
del july_2016_df['stop time']

#2019 Starttimes
july_2019_df['starttime'] = pd.to_datetime(july_2019_df['starttime'])
july_2019_df['start date'] = july_2019_df['starttime'].dt.date
july_2019_df['start time'] = july_2019_df['starttime'].dt.time
del july_2019_df['starttime']

#2019 Endtimes
july_2019_df['stoptime'] = pd.to_datetime(july_2019_df['stoptime'])
july_2019_df['end_date'] = july_2019_df['stoptime'].dt.date
july_2019_df['end_time'] = july_2019_df['stoptime'].dt.time
del july_2019_df['stoptime']

july_2019_df.head(10)
```

```python
#Change all spaces to underscores in the column namesto allow easy read to SQL
july_2016_df.columns = july_2016_df.columns.str.replace(' ','_')
july_2019_df.columns = july_2019_df.columns.str.replace(' ','_')

#Add an underscore to the tripduration, bikeid, and usertype
july_2019_df.rename(columns={'tripduration': 'trip_duration', 'bikeid': 'bike_id', 'usertype': 'user_type'}, inplace=True)
```

```python
#Create new files in the Resources folder for the cleaned data
july_2016_df.to_csv('Resources/Cleaned_july_2016_data.csv')
july_2019_df.to_csv('Resources/Cleaned_july_2019_data.csv')
```
