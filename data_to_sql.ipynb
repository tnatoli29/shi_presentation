```python
#Import Dependencies
import pandas as pd
from sqlalchemy import create_engine
```

```python
#Pull the cleaned data in
clean_2016 = pd.read_csv('Resources/Cleaned_july_2016_data.csv')
clean_2019 = pd.read_csv('Resources/Cleaned_july_2019_data.csv')

clean_2019.head()
```

```python
#Change the unnamed index column to index 
clean_2016.rename(columns = {'Unnamed: 0':'index'}, inplace = True)
clean_2019.rename(columns = {'Unnamed: 0':'index'}, inplace = True)
```

```python
#Create the connect between this page and Postgres
engine = create_engine(f'postgresql+psycopg2://tnatoli:postgres@localhost:5432/SHI_presentation')
conn = engine.connect()
```

```python
#Load the clean 2016 data onto postres
clean_2016.to_sql(name='citi_bike_2016', con=engine, if_exists='append', index=False)
```

```python
#Load the clean 2019 data onto postres
clean_2019.to_sql(name='citi_bike_2019', con=engine, if_exists='append', index=False)
```
